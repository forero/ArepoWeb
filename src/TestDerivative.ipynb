```python
x_line = random.random(100)*5.0*pi
x_line = sort(x_line)
```

```python
def my_func(x):
    return sin(x)
def my_func_prime(x):
    return cos(x)
```

```python
scatter(x_line, my_func(x_line))
```

```python
def approx_derivative(x, y, neigh=4):
    y_prime = y.copy()
    n_points = size(x)
    for i in range(n_points):
        delta = fabs(x[i]-x)
        sort_index = argsort(delta)
        tmp_x = x[sort_index[1:neigh+1]]
        tmp_y = y[sort_index[1:neigh+1]]
        
        tmp_prime = (tmp_y - y[i]) / (tmp_x - x[i])  
        y_prime[i] = average(tmp_prime)
    return y_prime
```

```python
a = approx_derivative(x_line, my_func(x_line), neigh=7)
```

```python
scatter(x_line, a)
plot(x_line, my_func_prime(x_line))
```

```python
plot(x_line, my_func_prime(x_line)-a)
```
