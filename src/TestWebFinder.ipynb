```python
def load_snap(path, snapnum, filenum):
    filename="%s/snap_%d.%d"%(path, snapnum, filenum)
    position = snap.read_block(filename, "POS ", parttype=0) 
    velocity = snap.read_block(filename, "VEL ", parttype=0)
    potential = snap.read_block(filename, "POT ", parttype=0)
    rho_gas = snap.read_block(filename, "RHO ", parttype=0)
    volume = snap.read_block(filename, "VOL ", parttype=0)
    return position, velocity, potential, rho_gas, volume
```

```python
snapnum=135
slice_x = empty((0))
slice_y = empty((0))
slice_z = empty((0))
slice_dens = empty((0))
slice_vx = empty((0))
slice_vy = empty((0))
slice_vz = empty((0))

data_path= "/Users/forero/work/ArepoWeb/data/L75n455FP/snapdir_135"

delta_slice = 500.0
center_slice = 30000.0
window_xy = 10000.0

for i in range(32):
    pos, vel, p, dens, vol = load_snap(data_path, snapnum, i)
    index = where((pos[:,2]>(center_slice-0.50*delta_slice)) & (pos[:,2]<(center_slice+0.50*delta_slice))&
                  (pos[:,0]>(center_slice-window_xy))&(pos[:,0]<(center_slice+window_xy))&
                  (pos[:,1]>(center_slice-window_xy))&(pos[:,1]<(center_slice+window_xy)))
    index = index[0]
    slice_x = append(slice_x, pos[index,0])
    slice_y = append(slice_y, pos[index,1])
    slice_z = append(slice_z, pos[index,2])
    slice_vx = append(slice_vx, vel[index,0])
    slice_vy = append(slice_vy, vel[index,1])
    slice_vz = append(slice_vz, vel[index,2])
    slice_dens = append(slice_dens, dens[index])
    print size(index)
print size(slice_x), size(slice_dens)
```

```python
index_center = where((slice_z>(center_slice-0.20*delta_slice))&(slice_z<(center_slice+0.20*delta_slice))&
                     (slice_x>(center_slice-0.8*window_xy))&(slice_x<(center_slice+0.8*window_xy))&
                      (slice_y>(center_slice-0.8*window_xy))&(slice_y<(center_slice+0.8*window_xy)))
                     
index_center = index_center[0]
#random.shuffle(index_center)

print size(index_center)
```

```python
MINR = 1.0
NMAX = size(index_center)
itera = 0 
neigh = 8
max_radius_neigh = empty((0))
min_radius_neigh = empty((0))
sigma_xx = empty((0))
sigma_yy = empty((0))
sigma_zz = empty((0))
sigma_xy = empty((0))
sigma_xz = empty((0))
sigma_yz = empty((0))

hubble = 100.0
for i in index_center[0:NMAX]:
    center_x = slice_x[i]
    center_y = slice_y[i]
    center_z = slice_z[i]
    center_vx = slice_vx[i]
    center_vy = slice_vy[i]
    center_vz = slice_vz[i]
    radius = sqrt((slice_x - center_x)**2 + (slice_y -center_y)**2 + (slice_z-center_z)**2)
  

   # print size(radius), slice_x[i-neigh*2:i+neigh*2], i-neigh*2, i+neigh*2, min_i, max_i
    rank_index = argsort(radius)
    max_radius_neigh = append(max_radius_neigh,radius[rank_index[neigh+1]])
    min_radius_neigh = append(min_radius_neigh,radius[rank_index[1]])

    inside = rank_index[1:neigh+1]
    
    tmp_dens = slice_dens[inside]
    tmp_vx = slice_vx[inside] 
    tmp_vy = slice_vy[inside] 
    tmp_vz = slice_vz[inside] 
    tmp_x = slice_x[inside]
    tmp_y = slice_y[inside]
    tmp_z = slice_z[inside]
    
    delta_x = tmp_x - center_x
    delta_x[abs(delta_x)<MINR] = MINR
    delta_y = tmp_y - center_y
    delta_y[abs(delta_y)<MINR] = MINR
    delta_z = tmp_z - center_z
    delta_z[abs(delta_z)<MINR] = MINR
    
    tmp_sigma_xx = -average((tmp_vx - center_vx)/(delta_x))/hubble
    tmp_sigma_yy = -average((tmp_vy - center_vy)/(delta_y))/hubble
    tmp_sigma_zz = -average((tmp_vz - center_vz)/(delta_z))/hubble
        
    tmp_sigma_xy = -0.5*average((tmp_vx - center_vx)/(delta_y) + (tmp_vy - center_vy)/(delta_x))/hubble
    tmp_sigma_xz = -0.5*average((tmp_vx - center_vx)/(delta_z) + (tmp_vz - center_vz)/(delta_x))/hubble
    tmp_sigma_yz = -0.5*average((tmp_vy - center_vy)/(delta_z) + (tmp_vz - center_vz)/(delta_y))/hubble
    
    sigma_xx = append(sigma_xx, tmp_sigma_xx)
    sigma_yy = append(sigma_yy, tmp_sigma_yy)
    sigma_zz = append(sigma_zz, tmp_sigma_zz)
    itera = itera+1
    if(not(itera%1000)):
         print itera, size(index_center) 
    
```
